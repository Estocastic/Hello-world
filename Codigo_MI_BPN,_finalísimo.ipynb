```python
# Supongamos que el DataFrame se llama df
categorical_columns = df_final3.select_dtypes(include=['object']).columns
selected_columns = [col for col in categorical_columns if "fecha" not in col.lower()]
selected_columns.append('def')  # Añadiendo la variable de interés

# Creando un nuevo DataFrame con solo las columnas seleccionadas
df_final3_selected =  df_final3[selected_columns]
```

```python
df_final3_selected.info()
```

```python
# Vamos a crear un diccionario para almacenar todas las tablas
all_tables = {}

# Lista de columnas categóricas para iterar y crear tablas
categorical_columns = df_final3_selected.columns[df_final3_selected.dtypes == 'object'].tolist()

# Para cada columna categórica, crear una tabla y agregarla al diccionario
for col in categorical_columns:
    # Agrupamos por la columna categorica y 'def'
    grouped = df_final3_selected.groupby(col)['def'].agg(['count', 'sum']).reset_index()
    grouped.columns = [col, 'Nacimientos', 'Muertes']

    # Calcular el TMI para cada categoría
    grouped['TMI'] = (grouped['Muertes'] / grouped['Nacimientos']) * 100

    # Agregar la tabla al diccionario
    all_tables[col] = grouped

# Mostrar todas las tablas creadas
print(all_tables)
```

```python
with open('/content/drive/My Drive/Datos_BPN2023/compiled_tables.csv', 'w') as outfile:
    for key, table in all_tables.items():
        # Escribir el nombre de la variable como cabecera de su tabla
        outfile.write(f"{key}\n")
        table.to_csv(outfile, sep=';', index=False, encoding = 'utf-8-sig')
        # Escribir una línea en blanco como separador entre tablas
        outfile.write("\n")


# Retornar la ruta al archivo compilado para el usuario
compiled_csv_path = '/content/drive/My Drive/Datos_BPN2023'
compiled_csv_path
```
